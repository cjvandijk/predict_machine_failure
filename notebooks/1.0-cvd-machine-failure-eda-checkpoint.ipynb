# Exploratory Data Analysis for Machine Failure data
Data source: [kaggle](https://www.kaggle.com/datasets/umerrtx/machine-failure-prediction-using-sensor-data?resource=download)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings('ignore')
```

```python
raw_data_file = "../data/raw/data.csv"
```

```python
df_raw = pd.read_csv(raw_data_file)
df_raw.head()
```

```python
df = df_raw.rename(columns={
    "tempMode":"temp_mode",
    "AQ":"air_quality", 
    "USS":"proximity_sensor", 
    "CS":"current_usage", 
    "VOC":"voc_level", 
    "RP":"revolutions_per_minute", 
    "IP":"input_pressure",
    "Temperature":"temperature"
})
```

```python
# Number observations, variables
df.shape
```

```python
# data types for each column (all columns are integer, 
# no need to separate out categorical features)

df.info()
```

```python
# How many null values are in dataset?
df.isnull().sum()
```

```python
# duplicated data?
df.duplicated().sum()
```

```python
# Distribution of target values looks good
fig = plt.figure(figsize = (3,2))
ax = fig.gca()

df.fail.hist(ax=ax)
```

```python
# check distribution of other column values

columns = list(df.columns)
columns.remove("fail")

for col in columns:
    sns.displot(df[[col]], rug=True, common_norm=False, kde=True, height=3)

plt.show()
```

```python
# Outliers: for each row, retain all column values that are within
# 3 standard deviations from the mean

df_orig = df.copy()

df = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)].copy()

print(f"Eliminated {len(df_orig) - len(df)} outliers")
```

```python
# Have a look at the outliers

df_outliers = pd.concat([df_orig, df]).drop_duplicates(keep=False)
df_outliers.describe()
```

```python
# Compare to the rows that were retained
df.describe()
```

```python
# Might we want to bin some column data?

for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

# Model Training

```python
X = df.drop('fail',axis=1)
Y = df['fail']

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2
)

len(X_train), len(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
lr.score(X_test, Y_test)
```
